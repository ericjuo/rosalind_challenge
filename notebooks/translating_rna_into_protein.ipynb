```python
import sys
sys.path.append("..")
from rosalind_tools.config import *
from Bio.Data import CodonTable
```

Given: An RNA string s corresponding to a strand of mRNA (of length at most 10 kbp).

Return: The protein string encoded by s.

```python
def translation(s: str) -> str:
    # Use codon table from biopython
    table = CodonTable.generic_by_id[1].forward_table
    stop_codons = CodonTable.generic_by_id[1].stop_codons
    protein_seq = ''
    for i in range(0, len(s), 3):
        if s[i:i+3] in stop_codons:
            break
        protein_seq += table[s[i:i+3]]
        i += 3
    print(protein_seq)
```

```python
# Try sample dataset
s = 'AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA'
translation(s)
```

```python
# Try Rosalind dataset
with open(data_dir/"rosalind_prot.txt", 'r') as f:
    s = f.read()
    translation(s)
```
